## 🔄 Python vs R: La Perspectiva General

Antes de escribir código, entendamos las diferencias de mentalidad:

| Aspecto | R | Python |
|---|---|---|
| Propósito original | Estadística y análisis | Programación general |
| Paradigma dominante | Funcional / vectorizado | Orientado a objetos |
| Sintaxis de bloques | `{ }` | Indentación obligatoria |
| Terminador de línea | Opcional (newline) | Newline (sin `;`) |
| Índices | Comienzan en **1** | Comienzan en **0** |
| Asignación | `<-` (preferido) o `=` | Solo `=` |
| NULL / nada | `NULL` | `None` |
| Valor verdadero | `TRUE` / `FALSE` | `True` / `False` |
| Comentarios | `#` | `#` ✅ (igual) |
| Documentación inline | Roxygen `#'` | Docstrings `"""` |

::: {.callout-tip title="💡 Mentalidad Clave"}
En R todo está vectorizado por defecto — operás sobre vectores completos sin pensar.
En Python el comportamiento por defecto es **escalar** (elemento a elemento). Para vectorización necesitás NumPy o pandas.
Esto no es mejor ni peor, es diferente. Lo veremos en profundidad en capítulos posteriores.
:::

### 1.2.2 `print()` en detalle

La función `print()` en Python es más explícita que en R. En R, escribir una expresión en la consola la imprime automáticamente. En Python **necesitás llamar `print()` explícitamente** (excepto en el intérprete interactivo).

## 1.3 Python como Calculadora

### 1.3.1 Operadores aritméticos básicos

````python
# Operaciones aritméticas básicas
# La sintaxis es idéntica a R en casi todos los casos

print("--- Operaciones básicas ---")
print(2 + 3)      # Suma          → 5
print(10 - 4)     # Resta         → 6
print(10 * 5)     # Multiplicación → 50
print(17 / 3)     # División real  → 5.666... (siempre retorna float)

print("\n--- Operadores exclusivos de Python ---")
print(17 // 3)    # División entera → 5   (en R: 17 %/% 3)
print(17 % 3)     # Módulo (resto)  → 2   (en R: 17 %% 3)
print(2 ** 10)    # Potencia        → 1024 (en R: 2^10 o 2**10)
````

::: {.callout-note title="R vs Python — Operadores"}
| Operación | R | Python |
|---|---|---|
| División real | `17 / 3` → 5.67 | `17 / 3` → 5.67 ✅ igual |
| División entera | `17 %/% 3` → 5 | `17 // 3` → 5 |
| Módulo | `17 %% 3` → 2 | `17 % 3` → 2 |
| Potencia | `2^10` o `2**10` | `2**10` (solo `**`) |

⚠️ En Python `^` es XOR (operación de bits), **no** potencia. ¡Cuidado!
:::

### 1.3.2 Tipos numéricos en Python

A diferencia de R donde casi todo es vector, Python tiene tipos escalares distintos:

````python
# Python tiene 3 tipos numéricos principales

entero    = 42          # int   — equivalente a integer en R
decimal   = 3.14        # float — equivalente a numeric/double en R
complejo  = 3 + 2j      # complex — equivalente a complex en R

print(type(entero))     # <class 'int'>
print(type(decimal))    # <class 'float'>
print(type(complejo))   # <class 'complex'>

# type() en Python == class() en R
print("\n--- El tipo determina el resultado ---")
print(10 / 2)     # 5.0  → float (aunque sea divisible exacto)
print(10 // 2)    # 5    → int
print(type(10 / 2))   # float
print(type(10 // 2))  # int
````

### 1.3.3 Precisión numérica y números grandes

Python maneja enteros de precisión arbitraria — algo que R no hace nativamente:

````python
# Python maneja enteros enormes sin overflow
numero_enorme = 2 ** 100
print(f"2^100 = {numero_enorme}")
print(f"Dígitos: {len(str(numero_enorme))}")

# En R, 2^100 retorna Inf si se usa integer, o notación científica en double
# Python: sin límite en enteros

print("\n--- Floats tienen precisión limitada (IEEE 754) ---")
print(0.1 + 0.2)           # 0.30000000000000004 — igual que R!
print(0.1 + 0.2 == 0.3)    # False — ¡trampa clásica!

# La solución correcta
import math
print(math.isclose(0.1 + 0.2, 0.3))   # True — equivalente a all.equal() en R
````

::: {.callout-warning title="⚠️ Trampa de Precisión Flotante"}
Este problema existe en **todos** los lenguajes (R, Python, JavaScript, C...).
La regla: **nunca compares floats con `==`**. Usá `math.isclose()` en Python o `all.equal()` en R.
:::

### 1.3.4 Operaciones combinadas y precedencia

````python
# La precedencia es la misma que en matemáticas (y en R)
# PEMDAS: Paréntesis, Exponentes, Multiplicación/División, Suma/Resta

resultado1 = 2 + 3 * 4        # 14, no 20
resultado2 = (2 + 3) * 4      # 20
resultado3 = 2 ** 3 ** 2      # 512 = 2^(3^2) = 2^9  (potencia: derecha a izquierda)

print(f"2 + 3 * 4     = {resultado1}")
print(f"(2 + 3) * 4   = {resultado2}")
print(f"2 ** 3 ** 2   = {resultado3}")

# Fórmula real: distancia euclidiana
x1, y1 = 0, 0
x2, y2 = 3, 4
distancia = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
print(f"\nDistancia entre (0,0) y (3,4): {distancia}")
````

### 1.4.2 Asignación múltiple y simultánea

Python tiene formas elegantes de asignar múltiples variables a la vez:

````python
# Asignación múltiple en una línea (unpacking)
x, y, z = 10, 20, 30
print(f"x={x}, y={y}, z={z}")

# Intercambiar valores — elegante en Python, sin variable temporal
a, b = 5, 8
print(f"Antes: a={a}, b={b}")
a, b = b, a       # Swap!
print(f"Después: a={a}, b={b}")

# Asignar el mismo valor a múltiples variables
p = q = r = 0
print(f"p={p}, q={q}, r={r}")
````

```r

a <- b
b <- temp
```
```python
# Python lo hace en una línea (tuple unpacking)
a, b = 5, 8
a, b = b, a    # ✨ más elegante
```

### 1.4.3 Convenciones de nombres

Python tiene convenciones estrictas (PEP 8):

````python
# ✅ Nombres válidos y convenciones PEP 8
nombre_usuario    = "carlos"      # snake_case — estándar en Python
edad_minima       = 18
MAX_INTENTOS      = 3             # CONSTANTES en mayúsculas
_variable_interna = "privada"     # _ al inicio = convención de "privado"

# ✅ Python distingue mayúsculas de minúsculas (como R)
radio = 5
Radio = 10
RADIO = 15
print(radio, Radio, RADIO)    # Son tres variables distintas

# ❌ Nombres inválidos (generan SyntaxError)
# 2variable = 10      # No puede empezar con número
# mi-variable = 10    # Guiones NO (solo guiones bajos _)
# for = 10            # Palabras reservadas del lenguaje
````

::: {.callout-note title="R vs Python — Convenciones de nombres"}
| Estilo | R | Python |
|---|---|---|
| snake_case | ✅ Común | ✅ Estándar (PEP 8) |
| camelCase | ✅ Común en R base | ⚠️ Solo para clases |
| punto.separado | ✅ Muy común en R | ❌ Evitar en Python |
| MAYUSCULAS | Variables de entorno | Constantes |

R acepta `mi.variable` (con punto). Python lo interpreta como acceso a atributo (`mi.variable` = atributo `variable` del objeto `mi`). **Nunca uses punto en nombres de variables Python.**

## 1.5 Comentarios y Documentación

### 1.5.1 Comentarios de una línea

````python
# Comentario de una línea — idéntico a R
x = 5   # También al final de una línea de código

# En Python NO existe /* comentario multilínea */ como en C/Java
# Se usan múltiples # líneas

# Buenos comentarios explican el POR QUÉ, no el QUÉ
# ❌ Malo:
x = x + 1    # suma 1 a x

# ✅ Bueno:
x = x + 1    # ajusta por índice base-0

print(x)
````

### 1.5.2 Strings multilínea como comentarios de bloque

````python
'''
Esto es un string multilínea con comillas simples.
Se usa como comentario de bloque por convención,
aunque técnicamente es un literal de string sin asignar.
El intérprete lo evalúa pero no hace nada con él.
'''

"""
Esta es la forma preferida para docstrings (documentación de funciones).
Las funciones en Python tienen su documentación dentro de ellas,
a diferencia de Roxygen en R que va ANTES de la función.
"""

# Ejemplo de función con docstring (adelanto del capítulo 3)
def calcular_area(radio):
    """
    Calcula el área de un círculo.

    Args:
        radio (float): El radio del círculo.

    Returns:
        float: El área del círculo.
    """
    return 3.14159 * radio ** 2

# Acceder a la documentación
help(calcular_area)
````

## 1.6 F-Strings: Interpolación de Variables

Las f-strings son el equivalente moderno de `paste()` / `sprintf()` en R:

````python
nombre = "Martín"
edad = 32
pi = 3.14159265

# F-string básica — prefijo f antes de las comillas
print(f"Hola, {nombre}! Tenés {edad} años.")

# Expresiones dentro de las llaves
print(f"El doble de tu edad es {edad * 2}")
print(f"En 10 años tendrás {edad + 10} años")

# Formato numérico
print(f"Pi redondeado a 2 decimales: {pi:.2f}")
print(f"Pi redondeado a 4 decimales: {pi:.4f}")
print(f"Número con separador de miles: {1234567891:,}")
print(f"Porcentaje: {0.856:.1%}")

# Alineación de texto
print(f"{'Izquierda':<20}|{'Centro':^10}|{'Derecha':>20}")
````

::: {.callout-note title="R vs Python — Interpolación de strings"}
```r
# R tiene varias opciones:
nombre <- "Martín"; pi_val <- 3.14159

paste("Hola,", nombre)                          # paste()
paste0("Hola, ", nombre, "!")                   # paste0() sin sep
sprintf("Pi = %.2f", pi_val)                    # sprintf() con formato
glue::glue("Hola, {nombre}!")                   # glue (más parecido a Python)
```
```python
# Python moderno usa f-strings (Python 3.6+)
f"Hola, {nombre}!"              # Lo más común hoy
f"Pi = {pi_val:.2f}"            # Con formato
```
Las f-strings de Python son muy similares al paquete `glue` de R.

## 1.7 Errores Comunes al Empezar

### 1.7.1 Error de indentación

````python
# ✅ Correcto — Python usa indentación como estructura
x = 10
if x > 5:
    print("Mayor que 5")   # 4 espacios (o 1 tab)
    print("Este también")  # Mismo nivel = mismo bloque
print("Esto está fuera del if")  # Sin indentación = fuera del bloque

# En R esto sería:
# if (x > 5) {
#   print("Mayor que 5")
# }
# La indentación en R es opcional (pero buena práctica)
````

````python
# Los errores más comunes — descomentá para ver el error

# 1. Usar ^ en lugar de ** para potencia
# resultado = 2^3    # Esto es XOR en Python, no potencia!
resultado = 2**3
print(f"2^3 = {resultado}")  # 8, correcto

# 2. Confundir True/False con TRUE/FALSE de R
# activo = TRUE    # NameError en Python
activo = True      # Correcto
print(f"Activo: {activo}")

# 3. Intentar concatenar string + número directamente
# print("Tengo" + 5 + "años")    # TypeError!
print("Tengo" + str(5) + " años")  # Correcto: convertir a str
print(f"Tengo {5} años")            # Mejor: usar f-string

# 4. División entera vs real (¡en R todo / es real!)
print(10 / 3)     # 3.333... (float)
print(10 // 3)    # 3 (int) — requiere // explícito
````

::: {.callout-warning title="🚨 Tabla de Errores Clásicos: R → Python"}
| Error | Código Erróneo | Código Correcto |
|---|---|---|
| Potencia | `2 ^ 3` | `2 ** 3` |
| Booleanos | `TRUE`, `FALSE` | `True`, `False` |
| Nada/null | `NULL` | `None` |
| Asignación | `x <- 5` | `x = 5` |
| Nombre con punto | `mi.var = 5` | `mi_var = 5` |
| String + número | `"edad: " + edad` | `f"edad: {edad}"` |
| Comparación float | `0.1 + 0.2 == 0.3` | `math.isclose(0.1+0.2, 0.3)` |
:::

## 🚀 Mini Proyecto: Reporte de Análisis Básico

````python
# Mini Proyecto: Reporte de Inversión Simple
# Simula el cálculo de interés compuesto y genera un reporte.
#
# Fórmula: A = P × (1 + r/n)^(n×t)
# Donde:
#   P = capital inicial
#   r = tasa anual (decimal)
#   n = número de veces que se compone por año
#   t = tiempo en años

# Datos de la inversión
nombre_inversor  = "Melanie Becerril"
capital_inicial  = 10000     # pesos/dólares
tasa_anual       = 0.08      # 8% anual
n_composicion    = 12        # mensual
anios             = 5

# Cálculo
monto_final = capital_inicial * (1 + tasa_anual / n_composicion) ** (n_composicion * anios)
ganancia    = monto_final - capital_inicial
rendimiento = (ganancia / capital_inicial) * 100

# Reporte
print("=" * 50)
print(f"{'REPORTE DE INVERSIÓN':^50}")
print("=" * 50)
print(f"{'Inversor:':<22} {nombre_inversor}")
print("-" * 50)
print(f"{'Capital inicial:':<22} ${capital_inicial:>12,.2f}")
print(f"{'Tasa anual:':<22} {tasa_anual*100:>11.1f}%")
print(f"{'Composición:':<22} {'Mensual':>12}")
print(f"{'Plazo:':<22} {anios:>7} años")
print("-" * 50)
print(f"{'Monto final:':<22} ${monto_final:>12,.2f}")
print(f"{'Ganancia:':<22} ${ganancia:>12,.2f}")
print(f"{'Rendimiento total:':<22} {rendimiento:>11.1f}%")
print("=" * 50)
````
